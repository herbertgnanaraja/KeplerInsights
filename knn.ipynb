```python
#Import CSV data

import numpy as np
import pandas as pd

pathToFile = r"dataset/"
fileName = 'dataset-reference.csv'
#fileName = 'cumulative_values.csv'

data = pd.read_csv(pathToFile+fileName, index_col=0)
pd.set_option('display.max_columns', None) # show all columns

data.head()
```

```python
#Skip colums that are not needed

skip_attr = [ 
    'koi_fittype', 
    'koi_limbdark_mod', 
    'koi_parm_prov', 
    'koi_tce_delivname',
    'koi_quarters',
    'koi_trans_mod',
    'koi_datalink_dvr',
    'koi_datalink_dvs',
    'koi_sparprov'
]

df = data.iloc[:, 14:] # remove first 13 columns of identification and target data (includes koi_fpflag_xx)
df.drop(skip_attr, axis=1, inplace=True) # drop columns with documentation data
df.shape # output is (9564, 59)
df.info()
```

```python
#Situation1: Drop rows with any missing values

df_clean = df.dropna(axis=1, how='all') # drop all NaN/null columns
df_clean = df_clean.loc[:, (df_clean.sum(axis=0) != 0)] # drop all columns with zero data
df_clean.shape # output is (9564, 51)
```

```python
df_clean.dropna(axis=0, how='any', inplace=True) # drop rows with any NaN/null values
df_clean.shape # output is (6682, 51)
```

```python
#Situation2a: Drop rows with threshold and impute missing values with 0

df_clean2a = df.dropna(axis=1, how='all') # drop all NaN/null columns
df_clean2a = df_clean2a.loc[:, (df_clean2a.sum(axis=0) != 0)] # drop all columns with zero data
df_clean.shape
```

```python
df_clean2a.shape
```

```python
df_clean2a.dropna(axis=0, thresh=len(df_clean2a.columns)*0.7, inplace=True) # drop rows with any NaN/null values @ 30% missing of 51 features
df_clean2a.shape
```

```python
df_clean2a.fillna(value = 0,
          inplace = True)
df_clean2a.info()
```

```python
#Situation2b: Drop rows with threshold and impute missing values with median

df_clean2b = df.dropna(axis=1, how='all') # drop all NaN/null columns
df_clean2b = df_clean2b.loc[:, (df_clean2b.sum(axis=0) != 0)] # drop all columns with zero data
df_clean.shape
```

```python
df_clean2b.dropna(axis=0, thresh=len(df_clean2b.columns)*0.7, inplace=True) # drop rows with any NaN/null values @ 30% missing of 51 features
df_clean2b.shape
```

```python
df_clean2b.fillna(df_clean2b.median(), inplace= True)
df_clean2b.info()
```

```python
#Situation3a: Fill missing values with zero

df_clean3a = df.dropna(axis=1, how='all') # drop all NaN/null columns
df_clean3a = df_clean3a.loc[:, (df_clean3a.sum(axis=0) != 0)] # drop all columns with zero data
df_clean3a.shape
```

```python
df_clean3a.fillna(value = 0,
          inplace = True)
df_clean3a.info()
```

```python
#Situation3b: Fill missing values with zero

df_clean3b = df.dropna(axis=1, how='all') # drop all NaN/null columns
df_clean3b = df_clean3b.loc[:, (df_clean3b.sum(axis=0) != 0)] # drop all columns with zero data
df_clean3b.shape
```

```python
df_clean3b.fillna(df.median(),
          inplace = True)
df_clean3b.info()
```

```python
#Adding target value

target_attr = 'koi_pdisposition'
#target_vals = {'CANDIDATE': 1, 'FALSE POSITIVE': 0}
df_clean[target_attr] = data[target_attr] # get prediction labels
df_clean2a[target_attr] = data[target_attr]
df_clean2b[target_attr] = data[target_attr]
df_clean3a[target_attr] = data[target_attr]
df_clean3b[target_attr] = data[target_attr]
```

```python
#Splitting X and y

from sklearn.model_selection import train_test_split, cross_validate, StratifiedKFold
X = df_clean.iloc[:, :-1]
y = df_clean[target_attr]

#X = scaled.iloc[:, :-1]
#y = scaled[target_attr]

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
# transform data
X_scaled = pd.DataFrame(scaler.fit_transform(X),columns = X.columns)

#print(df_clean.type)
scaled.head()

print('X:', X.shape)
print('y:', y.shape)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, stratify=y, shuffle=True)
```

```python
# Metrics and useful imports
from sklearn.metrics import accuracy_score, roc_auc_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.model_selection import train_test_split, cross_validate, StratifiedKFold

metric_names = ['Accuracy', 'Precision', 'Recall', 'F1', 'AUC']

def calc_metrics(model, X, y):
    y_hat = model.predict(X)
    metrics = [accuracy_score, precision_score, recall_score, f1_score, roc_auc_score]
    scores = [fn(y, y_hat) for fn in metrics]
    return scores
```

```python
#Linera Classifier
from sklearn import linear_model
clf = linear_model.SGDClassifier(max_iter=1000)
clf.fit(X_train, y_train)

for i, metric in enumerate(calc_metrics(clf, X_test, y_test)):
    print(f"{metric_names[i]:<9}: {metric:.3f}")

print("---------------------------------------------------------")

from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=2)
knn.fit(X_train, y_train)

for i, metric in enumerate(calc_metrics(knn, X_test, y_test)):
    print(f"{metric_names[i]:<9}: {metric:.3f}")
```
